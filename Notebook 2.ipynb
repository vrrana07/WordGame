```
#Vidhi Rana
#1/9/2023
#1.(Something that surprised me as I worked on my project was how much error handling I
#actually had to put into the code for a game, because user input can be very unpredictable,
#and I don't want the code to stop in the middle just because of a user error.)

#2.(I had challenges with using multiple functions and sometimes the syntax. First of all,
#using multiple functions that work together to run the game was hard. I had to make certain
#variables global, like game_type in order for each function to run without having an 
#unassigned variable. One of the biggest challenges was figuring out how to make just the 
#letter in user_guess show up in word_blank and it took me many tries to sequence each character correctly.)

#3.(I did end up asking my dad for help because originally, I didn't have any parameters
#and my attempt at doing so was unnsuccessful, so he helped me make game_type global, 
#so it could be used as an arguement. He gave me a tip to make detailed variables 
#instead of just 1 letter variables.)

#4.(I feel proud of my code because despite the lenth and difficulty, it was something 
#I enjoyed doing, and would do again, perhaps more complicated than this game. It feels
#great to click run and see your code work smooth, so I'm 
#glad I chose to create this game.) 


#5.(If I had more time for this project, I would love to add some graphics and turn 
#this into a hangman sort of game where there is a visual, so that people can 
#actually have fun playing the game.)



print("WORD GUESS")
print('INSTRUCTIONS: YOU WILL EITHER INPUT A RANDOM WORD INTO THE THE INPUT BELOW OR GENERATE A RANDOM WORD.\nTHEN YOU OR OTHERS WILL GUESS YOUR WORD.\nYOU WILL HAVE 5 STRIKES IF YOU GET THE WRONG LETTER.')

print(' __________                  __________' )
print('|  Custom  |                |   Word   |')
print('|   Word   |                | Generator|')
print('|____(A)___|                |____(B)___|')

def get_game_type(game_type):
    """

        Takes user input for which game mode to continue in.

        args:

            game_type: User input for game type a or b.

        returns:
            game type code
                1 (int): Random word generator option, continues in random_word. 
                or 2 (int): Own word input option, continues in enter_word

    """
#Takes previous user input, game_type and uses it to determine which value to return.
#If A is selected, 1 is returned. If B is selected, 2 is returned. Errors are handled
#accordingly. Return values guide what function will being next.
    x=0
    while x==0:
        if game_type=='A':
            print('Option `input your own word (A) selected`')
            x=1
            return 1
        elif game_type=='B':
            print('Option `generate a random word list (B) selected`')
            x=2
            return 2
        elif game_type== '':
            game_type=input('Do you want to input your own word or generate a random word list (pick A or B)?')
            game_type=game_type.upper()
        else:
            game_type=input('Do you want to input your own word or generate a random word list (pick A or B)?')
            game_type=game_type.upper()

def enter_word():
    """

        Gets user input for word choice, making sure there are no spaces.

        args:

        returns:

            guess_word: word input that will be guessed.

    """

#This gets input for the word that the user wants to guess. It uses a loop to make sure there are no spaces.
    x=1
    while x==1:
        #I had to figure out how to loop this so that there are no words with spaces inputed.
    
        if x==1:
            guess_word=input('Enter the word (no spaces): ').lower()
            spaces=' ' in guess_word
            #print(spaces)
            if spaces==True:
                x=1
            else:
                x=3
                return guess_word

def random_word():
    """

        Selects random word from word list to assign to guess_word

        args:


        returns:

            guess_word: word input that will be guessed.

    """
#This imports random and just imports a random word from the list and returns it.
    list=['Adult', 'Plane','Air', 'Carrier', 'Airforce', 'Airport', 'Alphabet', 'Baby', 'Backpack','Balloon','Banana','Bank','Barbecue','Cappuccino','Car','Carpet','Carrot','Drill','Ears','Earth','Electricity','Family','Fan','Feather','Festival','Girl','Gloves','Hat','Highway','Horoscope','Horse','Ice','Insect','Junk','Kitchen','Knife','Woman','Man','Leg','Library','Magnet','Map','Maze']
    import random
    return random.choice(list).lower()

    

def word_guess(game_type):
    """

        Main word_guess function, taking in user guesses.

        args:

            game_type: User input for game type a or b.

        returns: void

    """
    #This assigns whichever word was returned from either of the 2 function choices to a variable, guess_word.
    x=0 
    game_type=game_type.upper()
    x = get_game_type(game_type)
    guess_word = ''
    if x==1:
        guess_word = enter_word()

    if x==2:
        guess_word= random_word()
    
    print("Now guess the word :)")
    #print(guess_word)

#Creates word_blank,the main printed statement that will be used while the user guesses. Also creates
#Used_letters, the other printed statement for users to know what letters have already been guessed.
    chars=int(len(guess_word))
    word_blank='#'*chars
    print(word_blank)

    used_letters=['USED LETTERS: ',]
    lett=[]
    #lett is all letters in word
    for let in guess_word:
        lett.append(let)
    
#This while loop is to maintain the user guesses.
    y=0
    while y<5:
        occurs=[]
        #The if else statement stops the code when every letter is guessed, or continues until all strikes are used up.
        if word_blank.isalpha():
            print('You have guessed the word!')
            y=6

        else:
            z=0
            #Takes in user input and uses error handling to make sure it is a singular letter.
            user_guess=input('Choose a letter: ').lower()
            while z==0:
                if user_guess.isalpha()==False:
                    user_guess=input('Choose a LETTER not another character: ').lower()
                elif len(user_guess)>1:
                    user_guess=input('Only input 1 LETTER: ').lower()
                elif user_guess in used_letters:
                    user_guess=input('You have already inputted that value: ').lower()
                else:
                    z=1
            #This if- else statement checks if the guess was correct or not and gives a strike if incorrect.
            if user_guess in lett:
                print('Correct')
                #These for statements basically see how many occurances the correct guess has and uses some string addition to reveal it within word_blank in the correct position.
                for i in range(len(guess_word)):
                    if(guess_word[i]==user_guess):
                        occurs.append(str(i))
                        #print(occurs)
                for pos in occurs:
                    #print(length)
                    guess_string= word_blank[:int(pos)]+user_guess+word_blank[int(pos)+1:]
                    #This segment addes the guessed letter into used_letters, and doesn't let it repeat.
                    for c in guess_string:
                        if c.isalpha():
                            used_letters.insert(-1,c)
                        if c in used_letters:
                            repeat=used_letters.count(c)
                            if repeat>1:
                                used_letters.remove(c)
                    word_blank=guess_string 
                print(used_letters)
                print(guess_string)
                
            else:
                print('Try Again', user_guess, 'was not in the word!')
                y=y+1
                used_letters.append(user_guess)
                print(used_letters)
    
        
    print(guess_word.capitalize())

#This segment isn't a function, but controlls the amount of plays the player wants to have, so they can quit or continue playing.
p=0
while p==0:
    print(word_guess.__doc__)
    game=input("New Game? (y or n)").lower()
    if game=='y':
        game_type=input('Do you want to input your own word or generate a random word list (A or B)?')
        word_guess(game_type)
    elif game=='n':
        p=1
        print("bye.")
        break
    else:
        p=0
```
